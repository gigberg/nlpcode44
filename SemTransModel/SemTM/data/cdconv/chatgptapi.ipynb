````python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def create_binary_tree(depth, value=1):
    if depth == 0:
        return None
    node = Node(value)
    node.left = create_binary_tree(depth - 1, value * 2)
    node.right = create_binary_tree(depth - 1, value * 2 + 1)
    return node

# Create a binary tree with depth 5
tree_root = create_binary_tree(3)
````

````python
# 二叉树遍历函数
def traverse(root, level: int) -> None:
    if not root:
        return
    # 前序位置
    print(f"节点 {root} 在第 {level} 层")
    traverse(root.left, level + 1)
    traverse(root.right, level + 1)

# 这样调用
traverse(tree_root, 1)
````

````python
# 二叉树遍历函数
level = 1
def traverse(root) -> None:
    if not root:
        return
    # 前序位置
    global level
    print(f"节点 {root} 在第 {level} 层")

    level += 1
    traverse(root.left)
    traverse(root.right)
    level -= 1
    return

# 这样调用
traverse(tree_root)
````
